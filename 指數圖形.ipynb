# 2的指數成長圖

```python
def exponent(x: float) -> float:
    return 2 ** x
```

```python
data = [exponent(i) for i in range(10)]
data = list(enumerate(data))
```

```python
import matplotlib.pyplot as plt
```

```python
data_x = [i[0] for i in data]
```

```python
data_y = [i[1] for i in data]
```

```python
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/GenSenRounded-M.ttc',size=20)
```

```python
plt.plot(data_x, data_y)
plt.title("2的指數成長圖", fontproperties=font)
```

# 2的底數遞減圖

```python
def exponent(x: float) -> float:
    return 0.5 ** x
```

```python
data = [exponent(i) for i in range(10)]
data = list(enumerate(data))
```

```python
data_x = [i[0] for i in data]
data_y = [i[1] for i in data]
```

```python
plt.plot(data_x, data_y)
plt.title("2的底數遞減圖", fontproperties=font)
```
